## Problema 2:
Para una partícula de masa $m$ sometida a una fuerza de la forma $\vec{F}=-k\vec{r}$, se tiene que las soluciones a las ecuaciones de movimiento en el caso de 2 dimensiones, son de la siguiente forma:
\begin{align*}
    x(t) &= A_{x}\cos{(\omega t)}\\
    y(t) &= A_{y}\cos{(\omega t-\phi)}
\end{align*}
donde $\phi$ es la fase relativa de las oscilaciones $x$ y $y$.

Realice 3 gráficas en el plano $x-y$ variando las constantes $A_{x}$,$A_{y}$ en el notebook adjunto, para cada uno de los siguientes casos:

-   $\phi=0$
-   $\phi=\pi/2$
-   $\phi=\pi/4$

y explique el comportamiento en cada caso.

```python
# Funciones de «x» y «y» #
def x(time, amplitude=15, omega=3):
    global omega_x
    omega_x = omega
    return amplitude*np.cos(omega*time)

def y(time, amplitude=15, omega=3, phi=0):
    return amplitude*np.cos(omega*time-phi)

# Generate time values #
t = np.linspace(-10,10,1000);
```

```python
# Generando graficas
fig, axs = plt.subplots(1,3, sharey=True, figsize=(1.5*15,1.5*5))

# Especificaciones para las graficas
strphi = ['0','\pi/2','\pi/4']
colors = ['green','teal','navy']
angles = [0,np.pi/2,np.pi/4]
for i in range(3):
    # Graficas
    axs[i].plot(x(t),y(t,phi=angles[i]), color=colors[i])
    axs[i].set_title(f"$\phi = {strphi[i]}$")
    axs[i].grid(linestyle=(0, (3, 10, 1, 10)), linewidth=0.8, color='lightgray')

axs[0].set_ylabel('$y$');
plt.setp(axs[:], xlabel='$x$ axis label');
plt.savefig("pavercomova.jpg", dpi=150)
```

## Problema 3:
Para el oscilador anisotrópico bidimensional, las ecuaciones de movimiento son:
\begin{align*}
     x(t) &= A_{x}\cos{(\omega_{x} t)}\\
    y(t) &= A_{y}\cos{(\omega_{y} t-\phi)}
\end{align*}
donde $\phi$ es la fase relativa de las oscilaciones $x$ y $y$.

Realice 5 gráficas en el plano $x-y$ variando las constantes necesarias en el notebook adjunto, para distintos valores de $n_{x}$ y $n_{y}$ tales que $\omega_{x}/n_{x}=\omega_{y}/n_{y}$
y explique el comportamiento en cada caso.


```python
# Generando graficas
fig, axs = plt.subplots(5, sharex=True, figsize=(1.5*5,1.5*25))

# Especificaciones para las graficas
strphi = ['0','\pi/4','\pi/2','3\pi/4','\pi']
colors = ['green','turquoise','teal','navy','mediumblue']
nx = [7,9,3,8,2]
ny = [2,5,1,9,3]
angles = [0,np.pi/4,np.pi/2,3*np.pi/4,np.pi]
omx = 3
for i in range(5):
    # Graficas
    axs[i].plot(x(t,omega=omx),y(t,omega=omx*nx[i]/ny[i],phi=angles[i]), color=colors[i])
    axs[i].set_title(f"$\phi = {strphi[i]}$, $n_x/n_y={nx[i]}/{ny[i]}$")
    axs[i].grid(linestyle=(0, (3, 10, 1, 10)), linewidth=0.8, color='lightgray')

axs[-1].set_xlabel('$x$');
plt.setp(axs[:], ylabel='$y$ axis label');
plt.savefig("pavercomova2.jpg", dpi=150)
```
