```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
wy = 3
wx = 3

# Funciones de «x» y «y»
def x(time, amplitude=15, omega=wx):
    global Ax
    Ax = amplitude
    return amplitude*np.cos(omega*time)

def y(time, amplitude=15, omega=wy, phi=0):
    global Ay
    Ay = amplitude
    return amplitude*np.cos(omega*time-phi)
```

```python
# Generar valores para el tiempo
t = np.linspace(-10,10,1000);

# Valores para la fase relativa
phi = [0, np.pi/8, np.pi/4, np.pi/2, 3*np.pi/2]

# Valores para «n_x» y «n_y»
nx = [1,2,3,5,9]
ny = [9,7,5,3,4]
```

```python
# Generación de gráficas
fig, axs = plt.subplots(2,3, sharey=True, figsize=(15,10))

# Especificaciones para las gráficas
fase = ['0','\pi/8','\pi/4','\pi/2','3\pi/2']
colores = ['green', 'lightseagreen', 'teal', 'navy', 'steelblue']
for ax,i in zip(axs.flat,range(6)):
    if i==5:
        ax.axis('off')
    else:
        ax.plot(x(t,omega=(nx[i]/ny[i])*wy),y(t,phi = phi[i]), color=colores[i])
        ax.set_aspect('equal')
        ax.set_xlim(left=(Ax+1)*(-1), right=Ax+1)
        ax.set_ylim(bottom=(Ay+1)*(-1), top=Ay+1)
        ax.set_title(f"$\phi = {fase[i]}$")
        ax.grid(linestyle = '-.', linewidth=0.8, color='lightgray')

axs[0,2].set_xlabel("x")
plt.setp(axs[:,0], ylabel='$y$');
plt.setp(axs[-1,:], xlabel='$x$');
```
