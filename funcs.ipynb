# Hoja de Trabajo 9 
## Problema 2:
Para una partícula de masa $m$ sometida a una fuerza de la forma $\vec{F}=-k\vec{r}$, se tiene que las soluciones a las ecuaciones de movimiento en el caso de 2 dimensiones, son de la siguiente forma:
\begin{align*}
    x(t) &= A_{x}\cos{(\omega t)}\\
    y(t) &= A_{y}\cos{(\omega t-\phi)}
\end{align*}
donde $\phi$ es la fase relativa de las oscilaciones $x$ y $y$.

Realice 3 gráficas en el plano $x-y$ variando las constantes $A_{x}$,$A_{y}$ en el notebook adjunto, para cada uno de los siguientes casos:

-   $\phi=0$
-   $\phi=\pi/2$
-   $\phi=\pi/4$

y explique el comportamiento en cada caso.

```python
# Funciones de «x» y «y» #
def x(time, amplitude=15, omega=3):
    return amplitude*np.cos(omega*time)

def y(time, amplitude=15, omega=3, phi=0):
    return amplitude*np.cos(omega*time-phi)

# Generate time values #
t = np.linspace(-10,10,1000);
```

```python
# Generando graficas
fig, axs = plt.subplots(1,3, sharey=True, figsize=(1.5*15,1.5*5))

# Grafica 1
axs[0].plot(x(t),y(t,phi=0), color='green')

# Grafica 2
axs[1].plot(x(t),y(t,phi=np.pi/2), color='teal')

# Grafica 3
axs[2].plot(x(t),y(t,phi=np.pi/4), color='navy')

# Especificaciones para las graficas
strphi = ['0','\pi/2','\pi/4']
for i in range(3):
    axs[i].set_title(f"$\phi = {strphi[i]}$")
    axs[i].grid(linestyle=(0, (3, 10, 1, 10)), linewidth=0.8, color='lightgray')

axs[0].set_ylabel('$y$');
plt.setp(axs[:], xlabel='$x$ axis label');
plt.savefig("pavercomova.jpg", dpi=150)
```
